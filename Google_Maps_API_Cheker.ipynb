```python
import requests
import warnings
import json
import sys
import os
import pandas as pd

def scan_gmaps(apikey):
    results = []

    def check_api(url, api_name, cost):
        response = requests.get(url, verify=False)
        if response.status_code == 200 and b"error_message" not in response.content:
            results.append({"API Name": api_name, "Cost": cost, "Vulnerable": "Yes", "Details": url})
        else:
            reason = ""
            try:
                reason = response.json().get("error_message", str(response.content))
            except json.JSONDecodeError:
                reason = "Cannot parse response."
            results.append({"API Name": api_name, "Cost": cost, "Vulnerable": "No", "Details": reason})

    urls_to_test = [
        ("https://maps.googleapis.com/maps/api/staticmap?center=45%2C10&zoom=7&size=400x400&key=" + apikey, "Staticmap API", "$2 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/streetview?size=400x400&location=40.720032,-73.988354&fov=90&heading=235&pitch=10&key=" + apikey, "Streetview API", "$7 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/directions/json?origin=Disneyland&destination=Universal+Studios+Hollywood4&key=" + apikey, "Directions API", "$5 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/geocode/json?latlng=40,30&key=" + apikey, "Geocode API", "$5 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/distancematrix/json?units=imperial&origins=40.6655101,-73.89188969999998&destinations=40.6905615,-73.9976592&key=" + apikey, "Distance Matrix API", "$5 per 1000 elements"),
        ("https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input=Museum%20of%20Contemporary%20Art%20Australia&inputtype=textquery&fields=photos,formatted_address,name,rating,opening_hours,geometry&key=" + apikey, "Find Place From Text API", "$17 per 1000 elements"),
        ("https://maps.googleapis.com/maps/api/place/autocomplete/json?input=Bingh&types=%28cities%29&key=" + apikey, "Autocomplete API", "$2.83 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/elevation/json?locations=39.7391536,-104.9847034&key=" + apikey, "Elevation API", "$5 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/timezone/json?location=39.6034810,-119.6822510&timestamp=1331161200&key=" + apikey, "Timezone API", "$5 per 1000 requests"),
        ("https://roads.googleapis.com/v1/nearestRoads?points=60.170880,24.942795&key=" + apikey, "Nearest Roads API", "$10 per 1000 requests"),
        ("https://roads.googleapis.com/v1/snapToRoads?path=-35.27801,149.12958|-35.28032,149.12907&interpolate=true&key=" + apikey, "Snap to Roads API", "$10 per 1000 requests"),
        ("https://roads.googleapis.com/v1/speedLimits?path=38.75807927603043,-9.03741754643809&key=" + apikey, "Speed Limits API", "$20 per 1000 requests"),
        ("https://www.googleapis.com/geolocation/v1/geolocate?key=" + apikey, "Geolocation API", "$5 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/place/details/json?place_id=ChIJN1t_tDeuEmsRUsoyG83frY4&fields=name,rating,formatted_phone_number&key=" + apikey, "Place Details API", "$17 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=-33.8670522,151.1957362&radius=500&type=restaurant&key=" + apikey, "Nearby Search API", "$32 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/place/textsearch/json?query=restaurants+in+Sydney&key=" + apikey, "Text Search API", "$32 per 1000 requests"),
        ("https://maps.googleapis.com/maps/api/place/photo?maxwidth=400&photoreference=CnRnAAAA&key=" + apikey, "Places Photo API", "$7 per 1000 requests"),
    ]

    for url, api_name, cost in urls_to_test:
        check_api(url, api_name, cost)

    df = pd.DataFrame(results)
    return df

warnings.filterwarnings("ignore")

if __name__ == "__main__":
    if len(sys.argv) > 1:
        if sys.argv[1] in ["--api-key", "-a"]:
            if len(sys.argv) > 2:
                scan_gmaps(sys.argv[2])
            else:
                print("Missing API key, aborting.")
        elif sys.argv[1] in ["--help", "-h"]:
            print("Usage: python maps_api_scanner.py --api-key YOUR_API_KEY")
        else:
            print("Invalid arguments. Use --help for usage information.")
    else:
        apikey = input("Please enter the Google Maps API key you want to test: ")
        scan_gmaps(apikey)
```
